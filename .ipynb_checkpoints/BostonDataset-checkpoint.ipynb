```python
import pandas 
import numpy 
import pprint 
import matplotlib.pyplot as plot
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

src = load_boston()
target = src.target
df = pandas.DataFrame(data=src.data, columns=src.feature_names)
df.head()
```

```python
df.hist("CRIM")
df.hist("AGE")
```

```python
for i in df.columns:
    col = df[i].values.reshape(-1, 1)
    reg = LinearRegression().fit(col, target)
    prd = reg.predict(col) 
    plot.scatter(col, target)
    plot.scatter(col, prd)
    plot.title(i)
    plot.figure()
```

```python
print("interpretation of the results: ")
print("The target value of homes in town is lineary dependant on: ")
print("- CRIM - strongly negativily ")
print("- ZN - positively ")
print("- INDUS - negativily ")
print("- CHAS - positivly ")
print("- NOX - negativily ")
print("- RM - strongly positively ")
print("- AGE - weakly negativily ")
print("- DIS - weakly positively ")
print("- RAD - weakly negativily ")
print("- TAX - negativily ")
print("- PTRATIO - negativily ")
print("- B - positively ")
print("- LSTAT - strongly negativily")
```
