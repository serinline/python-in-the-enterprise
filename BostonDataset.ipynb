```python
import pandas 
import numpy 
import pprint 
import matplotlib.pyplot as plot
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

src = load_boston()
data = src.data
cols = src.feature_names
target = src.target


df = pandas.DataFrame(data=data, columns=cols)
df.head()
```

```python
df.hist(column="CRIM")
df.hist(column="AGE")
```

```python
values = {}
for i in df.columns:
    col = df[i].values.reshape(-1, 1)
    reg = LinearRegression().fit(col, target)
    prd = reg.predict(col)
    val = reg.score(col, target)
    values[i]=val

    plot.scatter(col, target)
    plot.scatter(col, prd)
    plot.title("{} -> {}".format(i, val))
    plot.figure()
    

for i in values:
    print(str(i) + ": " + str(values[i]))
```

```python
# interpretation of the results:
#     The target value of homes in town is lineary dependant on: 
#         - CRIM - strongly negativily
#         - ZN - positively
#         - NOX - negativily
#         - RM - strongly positively
#         - AGE - weakly negativily
#         - DIS - weakly positively
#         - RAD - weakly negativily
#         - TAX - negativily
#         - PTRATIO - negativily
#         - B - positively
#         - LSTAT - strongly negativily

# And it's not lineary dependant on for example CHAS, because it is binary value.
```
